# 1. 注意力分数

```python
import math
import torch
from torch import nn
from d2l import torch as d2l
```

```python
# 遮蔽softmax操作
def masked_softmax(X, valid_lens):
    """通过在最后一个轴上遮蔽元素来执行softmax操作"""
    if valid_lens is None:
        return nn.functional.softmax(X, dim=-1)
    else:
        shape = X.shape
        if valid_lens.dim() == 1:
            valid_lens = torch.repeat_interleave(valid_lens, shape[1])
        else:
            valid_lens = valid_lens.reshape(-1)
        X = d2l.sequence_mask(X.reshape(-1, shape[-1]), valid_lens, value=-1e6)   
        return nn.functional.softmax(X.reshape(shape), dim=-1)
```

- 老师：如果序列填充了东西，填充不算在做交叉熵的loss；但是不能把pad设为0，这样会导致exp出问题（e的0次方等于1，所以不能设成0），应该要把pad设为很小的数这样求exp才是0，这样算softmax的时候后面的东西就不会被算进来；

```python
# 演示此函数是如何工作
masked_softmax(torch.rand(2,2,4), torch.tensor([2,3]))
```

- 网友：参考注意力分数矩阵的形状是n*m 相当于一行对应一个query所以是对每行softmax 一些列mask掉舍去没用的填充
- 我：输入有两个batch，所以valmask的长度也是2；但是这里一个batch的所有样本的长度是统一的，不够细化；下面行代码可以把vallen细化到样本个体

```python
masked_softmax(torch.rand(2,2,4), torch.tensor([[1,3],[2,4]]))
```

- 我：这里在len长度为2的valmask中，再细化，增加内层列表，进一步表明每个batch的每个样本的长度

```python
# 加性注意力
class AdditiveAttention(nn.Module):
    """加性注意力"""
    def __init__(self, key_size, query_size, num_hiddens, dropout, **kwargs):
        super(AdditiveAttention, self).__init__(**kwargs)
        # 可以看到加性注意力机制实现需要：三个全连接层，一个dropout层
        # 这里bias都false，因为不需要
        self.W_k = nn.Linear(key_size, num_hiddens, bias=False)
        self.W_q = nn.Linear(query_size, num_hiddens, bias=False)
        self.w_v = nn.Linear(num_hiddens, 1, bias=False)
        self.dropout = nn.Dropout(dropout)# 可以正则化
      
    # valid_lens：可以把pad出来的key和value丢掉
    def forward(self, queries, keys, values, valid_lens):
        queries, keys = self.W_q(queries), self.W_k(keys)
        # 网友：这里用到了torch加法的广播机制
        features = queries.unsqueeze(2) + keys.unsqueeze(1)
        features = torch.tanh(features)
        scores = self.w_v(features).squeeze(-1)
        self.attention_weights = masked_softmax(scores, valid_lens)
        # 对哪些keyvaluepair不需要看，就drop掉
        # 网友：前者形状和score相同,然后由于是按batch做乘法,所以返回值形状就是 (batch_size, 查询的个数, 值的维度)
        return torch.bmm(self.dropout(self.attention_weights), values)
```

- 老师：init函数不需要知道valuesize，因为可学参数与它无关
- 老师：forward函数的queries维度：（batchsize,querysize,numhidden）
    - 网友：自己Print了一下，第二维是时间步，根本不是query_size
- 我：kq相加后计算每个k对应的权重，权重乘以value得到加性注意力机制

```python
# 演示上面的AdditiveAttention类
# (2,1,20)：batchsize是2,1个query，query长度是20。(2,10,2)：batchsize是2,10个key，key的长度是2
queries, keys = torch.normal(0, 1, (2,1,20)), torch.ones((2,10,2))
# (1,10,4)：1会因为repeat而变为2；10个value；value的长度是4
values = torch.arange(40, dtype=torch.float32).reshape(1,10,4).repeat(2,1,1)  
# 第一个样本查query的时候去看前2个kv对，第二个样本查query的时候去看前6个kv对；因为qkv的batchsize都是2，（每个batch都是一个样本），所以一共两个样本
valid_lens = torch.tensor([2,6])
attention = AdditiveAttention(key_size=2, query_size=20, num_hiddens=8, dropout=0.1)  
attention.eval()
attention(queries, keys, values, valid_lens)
```

```python
# 注意力权重
d2l.show_heatmaps(attention.attention_weights.reshape((1,1,2,10)),
                 xlabel='Keys', ylabel='Queries')
```

```python
# 缩放点积注意力
class DotProductAttention(nn.Module):
    """缩放点积注意力"""
    def __init__(self, dropout, **kwargs):
        super(DotProductAttention, self).__init__(**kwargs)
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, queries, keys, values, valid_lens=None):
        d = queries.shape[-1]
        scores = torch.bmm(queries, keys.transpose(1,2)) / math.sqrt(d)
        self.attention_weights = masked_softmax(scores, valid_lens)
        return torch.bmm(self.dropout(self.attention_weights), values)
```

```python
# 演示上述的DotProductAttention类
queries = torch.normal(0,1,(2,1,2))
attention = DotProductAttention(dropout=0.5)
attention.eval()
attention(queries, keys, values, valid_lens)
```

```python
# 均匀的注意力权重
d2l.show_heatmaps(attention.attention_weights.reshape((1,1,2,10)),
                 xlabel='Keys', ylabel='Queries')
```

- 老师：点乘attention实现比较简单，加性attention实现比较困难

1，核回归里面算法后面的那个红色的热图是什么意思?
- 老师：就是把注意力权重画出来；任何一行表示给每个kv对多少注意力；即query对哪个kv对觉得重要一点，这个就是weight
- 网友：以回归例子，训练集的数据x是key，越红色的地方说明输入query越关注对应的key

2，q k v的主要区别是？
- 老师：先不剧透，后面说应用会讲

3，把之前的x和x_i换成了q和k_i吗，这两个值是哪里的呢?
- 老师：对的，核回归的时候用的是x和xi，讲向量版本的时候用的是q和ki

4，为什么注意力可以理解为两者的相似度
- 老师：不是说理解为相似度，而是通常用相似度做注意力的分数；要查询一个query的值的话，一堆值在那，那我就会看谁的key和我相近，那么它的value就会比较重要
- 老师和网友：k是公司员工的能力，v是公司员工的薪水，q是你自己的能力，通过注意力可以大致算出你将来的薪水；给和自己条件差不多的人的权重高，条件差距大的人的权重低

5，maskedrsoftmax()就相当于把注意力分数低的过滤掉吗?那多低算低呢，这个阈值怎么设置的呢?
- 老师：举例子：一句话里有四个词，如果四个词不够一个句子的长度（10），就要pad6个pad符号；现在拿一个query过来，要查询怎么翻译某个字的话，那么payattention的时候10个词就是10个kv对，那么就应该只看前四个不看后六个，因为知道后面六个kv对是没意义的pad；masksoftmax就是算softmax的时候只考虑非pad部分的值；等于把后面pad的值的softmax都设成0。

# 1. 注意力机制

# 重点知识
1，
- 非参就是不需要学习参数

```python
import torch
from torch import nn
from d2l import torch as d2l
```

```python
# 生成数据集
n_train = 50
# 在0-5之间随机生成50个数，然后排序
x_train, _ = torch.sort(torch.rand(n_train) * 5)

def f(x):
    # f(x)=2sinx+x^0.8
    return 2 * torch.sin(x) + x**0.8

# fx加上方差为0.5的噪音，作为用于训练的label
y_train = f(x_train) + torch.normal(0.0, 0.5, (n_train,))
# 用于测试的x点
x_test = torch.arange(0, 5, 0.1)
# 测试点的真实y
y_truth = f(x_test)
n_test = len(x_test)
n_test
```

```python
def plot_kernel_reg(y_hat):
    d2l.plot(x_test, [y_truth, y_hat], 'x', 'y', legend=['Truth','Pred'],
            xlim=[0,5], ylim=[-1,5])
    d2l.plt.plot(x_train, y_train, 'o', alpha=0.5)
    
y_hat = torch.repeat_interleave(y_train.mean(), n_test)
# 任务就是给定xtest来尝试预测ytruth
plot_kernel_reg(y_hat)
```

- 老师：如果根据train直接做平均，那么得到pred，这个truth的形状相差较大。所以下面引入非参数注意力汇聚。

```python
# 非参数注意力汇聚
X_repeat = x_test.repeat_interleave(n_train).reshape((-1, n_train))
attention_weights = nn.functional.softmax(-(X_repeat - x_train)**2 / 2, dim=1)  
y_hat = torch.matmul(attention_weights, y_train)
plot_kernel_reg(y_hat)
```

- 网友：就是计算与目标点的x距离，距离近的就认为相似，然后用各个点的加权求和对目标点进行更新
- 老师：非参数的好处：
    - 不需要学
    - 有理论证明，就是只要给足够多的数据，就可以把原始的模型弄出来；但是现实生活中没那么多数据，所以非参数在机器学习用的不多

```python
# 注意力权重
d2l.show_heatmaps(attention_weights.unsqueeze(0).unsqueeze(0),
                xlabel='Sorted training inputs', ylabel='Sorted test inputs')   
```

```python
# 带参数注意力汇聚
# 假定两个张量的形状分别是(n,a,b)和(n,b,c)，它们的批量矩阵乘法输出的形状为(n.a,c)  
X = torch.ones((2,1,4))
Y = torch.ones((2,4,6))
torch.bmm(X, Y).shape# bmm是批量矩阵乘法：1行四列的矩阵，乘以4行6列的矩阵，得到的就是1行六列的矩阵
```

```python
# 使用小批量矩阵乘法来计算小批量数据中的加权平均值
weights = torch.ones((2,10)) * 0.1
values = torch.arange(20.0).reshape((2,10))
torch.bmm(weights.unsqueeze(1), values.unsqueeze(-1))
```

```python
# 带参数的注意力汇聚
class NWKernelRegression(nn.Module):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.w = nn.Parameter(torch.rand((1,),requires_grad=True))  
        
    def forward(self, queries, keys, values):
        queries = queries.repeat_interleave(keys.shape[1]).reshape(-1,keys.shape[1])   
        # w可以控制曲线的平滑度，通过学习w可以控制曲线平滑or不平滑
        self.attention_weights = nn.functional.softmax(-((queries - keys) * self.w)**2/2,dim=1)     
        return torch.bmm(self.attention_weights.unsqueeze(1),values.unsqueeze(-1)).reshape(-1)   
```

```python
# 将训练数据集转换为键和值
X_tile = x_train.repeat((n_train, 1))
Y_tile = y_train.repeat((n_train, 1))
keys = X_tile[(1 - torch.eye(n_train)).type(torch.bool)].reshape((n_train,-1))  
values = Y_tile[(1 - torch.eye(n_train)).type(torch.bool)].reshape(n_train, -1) 
```

```python
# 训练带参数的注意力汇聚模型
net = NWKernelRegression()
loss = nn.MSELoss(reduction='none')
trainer = torch.optim.SGD(net.parameters(), lr=0.5)
animator = d2l.Animator(xlabel='epoch',ylabel='loss',xlim=[1,5])

for epoch in range(5):
    trainer.zero_grad()
    l = loss(net(x_train, keys, values), y_train) / 2
    l.sum().backward()
    trainer.step()
    print(f'epoch {epoch+1}, loss {float(l.sum()):.6f}')
    animator.add(epoch+1, float(l.sum()))
```

```python
# 预测结果绘制
keys = x_train.repeat((n_test, 1))
values = y_train.repeat((n_test, 1))
y_hat = net(x_test, keys, values).unsqueeze(1).detach()
plot_kernel_reg(y_hat)
```

- 老师：数据少，为了拟合数据需要窄的窗口，所以预测曲线不平滑；不那么平滑的好处就是红线接近蓝线，坏处就是抖动比较多。

```python
# 曲线在注意力权重较大的区域变得更不平滑
d2l.show_heatmaps(net.attention_weights.unsqueeze(0).unsqueeze(0),
                 xlabel='Sorted training inputs', ylabel='Sorted testing inputs')   
```

- 老师：可以看到红线比较窄；现在和x比较接近的xi给了比较大的权重，别的都给的比较小
- 网友：就是训练样本里面和测试样本靠近的点对预测结果的权重大
- 网友：横着看，一行的宽度明显窄于之前的曲线，证明只在该点附近的权重较大，较远的点的权重小或取0

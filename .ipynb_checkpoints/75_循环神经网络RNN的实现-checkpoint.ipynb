```python
%matplotlib inline
import math
import torch
from torch import nn
from torch.nn import functional as F
from d2l import torch as d2l

# 一个batch有多少样本，一个样本的字符数
batch_size, num_steps = 32, 35
train_iter, vocab = d2l.load_data_time_machine(batch_size, num_steps)
```

```python
# 独热编码
print(len(vocab))
# 生成两个独热编码，第一个在0号位置为1，第二个在2号位置为1
F.one_hot(torch.tensor([0,2]),len(vocab)) # [0,2] 表示物体下标，0表示第一个元素，2表示第3个元素
```

```python
# 小批量形状是(批量大小，时间步数)
X = torch.arange(10).reshape((2,5)) # 批量大小为2，时间步数为5
# 转置是为了让时间步数对应的维度变化到批量大小的前面。每次访问Xt都是连续的，因为挪掉了后面的维度
F.one_hot(X.T, 28).shape
```

- 网友：转置的作用是方便访问批量数据里所有某一时刻x(t)的数据

```python
# 初始化循环神经网络模型的模型参数
# device是为了适配GPU
def get_params(vocab_size, num_hiddens, device):
    # 因为类别的个数就是vocab_size，因为下一个词可以是字典vocab里的任何一个词。这里可以看到输出层28个点位其实就是一个字符的onehot
    num_inputs = num_outputs = vocab_size
    
    # 因为要不停初始化参数，所以要构建一个这个函数
    def normal(shape):
        # torch.randn(size=shape, device=device)：生成均值为0，方差为1的随机初始化权重；整体乘以0.01后，方差变成0.0001，均值还是0
        return torch.randn(size=shape, device=device) * 0.01
    
    W_xh = normal((num_inputs, num_hiddens))# 输入层到隐藏层的权重
    # 注意！！！：把下面这行移除后，就是单隐藏层的MLP
    W_hh = normal((num_hiddens, num_hiddens))# 上一个隐藏层到下一个隐藏层之间的权重
    b_h = torch.zeros(num_hiddens, device=device)# 每个隐藏层的初始化偏置bias
    W_hq = normal((num_hiddens, num_outputs))# 隐藏层到输出层的w
    b_q = torch.zeros(num_outputs, device=device)# 输出层的初始化偏置
    # 把参数都放到一个列表里，方便返回
    params = [W_xh, W_hh, b_h, W_hq, b_q]
    for param in params:
        # 对每个参数都需要算梯度，因为需要更新
        param.requires_grad_(True)
    return params
```

- 网友：一个input就只是特征向量的一个维度的独热编码串而已。。
    - 我：应该就是一个字符的onehot
- 我：要理解输入输出和权重的维度，可以[参考](http://t.csdn.cn/7Xb3Z)和[参考](https://zhuanlan.zhihu.com/p/460282865)；主要抓住对于**单个时间步**来说输入维度是（batch_size, word_features），W_hq = normal((num_hiddens, num_outputs))即每一个隐藏层单元都要参与一个输出;这一个输出的维度是numoutput即一个onehot。 
- 我：理解输入输出的维度还可以[！！！重点参考](https://www.zhihu.com/question/268862438/answer/346187736)，查看评论可知一个rnn的单元就是一个全连接层；了解hidden_size，循环步长num_steps，词向量维度embed_dim的关系。

```python
# 一个init_rnn_state函数在初始化时返回隐藏状态。之后会把函数名里的RNN换成GRU和LSTM
def init_rnn_state(batch_size, num_hiddens, device):
    return (torch.zeros((batch_size, num_hiddens), device=device),)
```

- 老师：因为在初始0时刻，没有上一时刻的隐藏状态，所以要给一个初始的隐藏状态，这样可以一直更新下去
- 老师：对每一个样本，他的隐藏状态是一个长为num_hidden的向量；所以有batchsize个样本的话，就有(batch_size, num_hiddens)形状的参数
    - 我问问问：训练好一个样本后，隐藏状态就会变，那么下一个样本进来就已经有隐藏状态了；所以难道不是第一个样本有隐藏状态就可以吗？
        - 我答：首先，隐藏状态只在一条sample有用，不同sample是不一定有时序关系的，所以需要有batchsize的维度；num_hiddens对每一个隐藏位置都要决定对后面的每个隐藏位置是否有联系，所以要初始化
- 老师：这里可以用0来初始化，或者用随机数，都可以
- 老师：可以看到返回的是一个tuple，因为RNN的隐藏状态就是一个张量就没了，但是之后的lstm的隐藏状态有两个东西；所以为了统一化，就把隐藏状态做成tuple；这里虽然只有一个，但是之后会有两个，所以写起来会比较方便。

```python
# 下面的rnn函数定义了如何在一个时间步计算隐藏状态和输出
def rnn(inputs, state, params):
    # param展开
    W_xh, W_hh, b_h, W_hq, b_q = params
    # state就是一个tuple，但是它目前只有一个元素。网友：这里加一个逗号是为了对tuple做解包，不加的话H就变成tuple类型了
    H, = state
    outputs = []
    # 沿着input的第一个维度去遍历；比如首先拿到的是时刻0对应的x，每个x的尺寸是批量大小*vocabsize；接下来就拿时刻1..一直到时刻t；每一次迭代就是算一个特定的时间步
    for X in inputs:
        # 当前时刻先更新隐藏状态：这里激活函数用了tanh
        H = torch.tanh(torch.mm(X, W_xh) # 输入X和Wxh做矩阵乘法
                       + torch.mm(H, W_hh) 
                       + b_h)
        # y是当前时刻的预测，也就是说在当前时刻预测下一时刻的词是谁
        Y = torch.mm(H, W_hq) + b_q
        # 因为是forloop，所以要把所有时刻的输出都append
        outputs.append(Y)
    '''
    # 函数返回
    1，output：每个时刻尺寸是batchsize*vocabsize
    2，cat是拼接：在0维度拼接（n个矩阵按照垂直方向拼起来），拼出来是二维矩阵；矩阵列数没变还是vocabsize，矩阵行数变为了batchsize*时间序列长度
    - 网友：用pycharm debug跟踪下更好理解
    3，(H,)：要输出更新后的隐藏状态，
    4，网友：从这里也能看出上节课有问题了，代码上显示的是当前的输出取决于当前的隐藏状态和当前的输入，而不是上节课所讲的上一个时刻的输入
    '''
    return torch.cat(outputs, dim=0), (H,)
```

- 老师：这个函数就计算，给一个小批量，然后把rnn的所有时间步都算一遍，然后得到输出
- 老师：参数解释：
    - input包含所有的时间步，从x0到xt
    - state就是初始化的隐藏状态
    - param就是可学习的参数
- 老师：input解释：
    - 维度：参考in[3]，shape=(时间序列的长度(即时间的步数，批量大小，vocabsize)，是一个3d的tensor
- 网友问：这里一直强调同一时刻有批量大小，应该是同一时刻不止一个输入？
    - 我：感觉是的，一个batch同时计算某一时刻（可能torch有并行优化，这样就是批量处理能加速计算）
- 老师：这里input每次拿一个时刻，是因为之前in[3]的X转置了，把时刻维度放在最前面才能做到
- 老师：这个rnn函数有点像mlp的forward函数，但是有几个区别：
    - input里面不再是批量大小*样本，还加上了时间维度
    - 参数有state：我们既要输入隐藏状态，也要输出隐藏状态

```python
# 创建一个类来包装这些函数
class RNNModelScratch:
    def __init__(self, vocab_size, num_hiddens, device, get_params,
                init_state, forward_fn):
        self.vocab_size, self.num_hiddens = vocab_size, num_hiddens
        self.params = get_params(vocab_size, num_hiddens, device)
        # 返回：初始的隐藏状态，和RNN函数；因为后面要通用GRU和LSTM，所以这里做的通用一点
        self.init_state, self.forward_fn = init_state, forward_fn
    
    # 这个X和之前的input不一样，这里的X就是《时间机器》加载进来后的X；尺寸是“批量大小*时间步数”，时间步数就是一句话的长度
    def __call__(self, X, state):
        # X被转化为的尺寸：时间步数*批量大小*vocabsize
        X = F.one_hot(X.T, self.vocab_size).type(torch.float32)
        # 把X放到RNN函数（现在叫forward_fn），可以得到输出和更新后的状态
        return self.forward_fn(X, state, self.params)
    
    def begin_state(self, batch_size, device):
        # 用initstat函数获取初始状态
        return self.init_state(batch_size, self.num_hiddens, device)
```

```python
- 老师：模型做forward有两种方式：
    - call一个forward函数
    - 自定义python里的call
```

```python
# 检查输出是否具有正确的形状
num_hiddens = 512
# 这里参数的用法，可以参考RNNModelScratch
net = RNNModelScratch(len(vocab), num_hiddens, d2l.try_gpu(), get_params,
                     init_rnn_state, rnn)
# 初始化隐藏状态，即时刻0之前的隐藏状态
state = net.begin_state(X.shape[0], d2l.try_gpu())
# X是2*5的刚刚定义的矩阵；to是放到GPU上。结合输入的旧隐藏层状态，得到返回的“输出+新隐藏层状态”
Y, new_state = net(X.to(d2l.try_gpu()),state)
Y.shape, len(new_state), new_state[0].shape
```

- 老师：现在看看拿到东西的形状是不是我们想要的
- 老师：Y的形状：第一个维度是batchsize*timestep；第二个维度是对于X中的每一个词，它的下一个词的预测向量，是一个长为28（26英文字母+空格+unk）的向量。说白了就是对X中的每一个词，去预测X的下一个词是谁；因为X有2*5=10个词，所以Y就是对10个词，每个都做预测，预测的结果是一个28长度的onehot向量。
    - ！！！注意我：执行一次net，就是把所有数据使用一遍，也就会针对数据里的每个字符去预测下一个时间步的字符；至于具体是怎么预测的，可以看RNN函数，它里面是按照时间步从小到大去做预测，所以靠后的时间步能够使用靠前时间步的“隐藏信息+输入”
- 老师：new_state的形状：它就是长为1的tuple

    - 我：突然想到隐藏状态为什么需要分为不同的batchsize；因为隐藏状态是服务于一条输入（输入的长度就是timestep）的，一条样本的不同字符（即上下时间步）之间会有依赖关系，但是不同样本的字符之间是没有依赖关系的；所以针对每个样本都要有隐藏层状态，所以隐藏层参数需要有batchsize的维度（batchsize就是样本的个数）
    - 我：隐藏状态只在当前样本有用，不同样本能共同使用的是层之间的权重值

```python
# 首先定义预测函数来生成用户提供的prefix之后的新字符
def predict_ch8(prefix, num_preds, net, vocab, device):
    """在 'prefix' 后面生成新字符。"""
    # 生成初始的隐藏状态。batch_size==1是因为对一个字符串做预测；
    state = net.begin_state(batch_size=1, device=device)
    '''
    1，网友：这里vocab不是词频表了！仅仅是一个字典map！把token映射出一个数
    2，老师：把第一个字符拿出来，映射为整数下标，然后放到output里面；所以output一开始只有一个字符
    '''
    outputs = [vocab[prefix[0]]]
    '''
    # 是一个函数，作用是把最近预测出的那个词，作为下一时刻的输入；
    1，reshape是实现批量大小是1，时间步长也是1，所以getinput是1*1的矩阵。因为是针对单个字符串prefix预测，所以预测是batch为1；
    2，为了输入给模型，所以必须有batch这个维度，所以有reshape
    '''
    get_input = lambda: torch.tensor([outputs[-1]],device=device).reshape(1,1)  
    # 给的prefix的0位置的字符已经放到output里，所以对后面的词都要遍历
    for y in prefix[1:]:
        '''
        1，老师：这里每次都会输出一个字符和隐藏层状态
        2，老师：但是我们不关心输出，因为这里会把真实的prefix放入output避免误差，我们想要的只是把prefix中的信息保存到状态state中
        '''
        _, state = net(get_input(), state)
        # 可以看到在处理prefix阶段的预测时，output存的都是真实的prefix而不是预测的结果
        outputs.append(vocab[y])
    # 把prefix所有的信息都存入state后，就可以真正预测了。需要预测num_preds个字符
    for _ in range(num_preds):
        '''
        # 每次都是把前一时刻的输出拿来作为输入
        1，y是1*vocabsize的向量
        '''
        y, state = net(get_input(), state)
        '''
        # 把y中值最大的dimension拿出来，reshape为一个标量，然后转化为一个int并存入output
        1，这个东西就是给定当前词和之前看到的东西，预测下一个词对应的下标，并把下标放入output
        '''
        outputs.append(int(y.argmax(dim=1).reshape(1)))
    # 把output中的每个index转为token并塞入列表，在把列表的元素合并起来
    return ''.join([vocab.idx_to_token[i] for i in outputs])

'''
# 给定《时间机器》里最常出现的“time traveller”作为prefix，来预测后面的十个字符
1，网友：模型还没训练，现在预测的结果应该是乱码
'''
predict_ch8('time traveller ', 10, net, vocab, d2l.try_gpu()) # 还没训练，这是随即初始化后的预测 
```

- 老师函数参数讲解：
    - prefix：给一个句子的开头prefix，然后生成接下来的句子
    - num_preds：要生成多少个字符；这里vocab用的是字符，所以这里词和字符没区别就是一个char
    - net：训练好的模型
    - vocab：可以把数值型的预测词映射成真实的字符串词
    - device：GPU或者CPU
- 网友：先讲预测，再讲训练

① 梯度裁剪$$\mathbf{g} \leftarrow \min\left(1, \frac{\theta}{\|\mathbf{g}\|}\right) \mathbf{g}$$

```python
# 这是做梯度剪裁的辅助函数
def grad_clipping(net, theta):
    """裁剪梯度。"""
    '''
    # 如果是nnmodule的话，从nnmodule中拿到param；不然刚刚实现的函数params就能拿到参数
    1，注意：这里是吧所有可以训练的层都拿出来，而不是一层层拿；p.requires_grad表示可以梯度下降，即可以训练
    '''
    if isinstance(net, nn.Module):
        params = [p for p in net.parameters() if p.requires_grad]
    else:
        params = net.params
    
    #L2norm正则化
    norm = torch.sqrt(sum(torch.sum((p.grad**2)) for p in params))
    '''
    # 如果norm大于theta，就把梯度映射回去；梯度映射回去后，再求梯度的norm就不会比theta大了
    1，网友：原来是一起裁剪....
    2，网友：所有参数限制在半径theta的高维球内
    '''
    if norm > theta:
        for param in params:
            param.grad[:] *= theta / norm
```

- 老师：RNN给一个小批量进去，就算批量里只有一个样本，并且RNN至于一个单隐藏层；但是单隐藏层要做T（时间步数即句子长度即设为35）步迭代，所以等效于有35层的MLP，有至少35个矩阵乘法在里面，所以容易有梯度爆炸；梯度爆炸的一个解决办法就是把梯度映射回去（公式之前讲过）

```python
# 定义一个函数来训练只有一个迭代周期的模型。ch8就是第八章的意思。
def train_epoch_ch8(net, train_iter, loss, updater, device, use_random_iter):
    """训练模型一个迭代周期"""
    # state初始化为none，timer是计时器
    state, timer = None, d2l.Timer()
    # 用来存各种loss信息
    metric = d2l.Accumulator(2)
    for X, Y in train_iter:
        if state is None or use_random_iter:
            '''
            1，网友问：不过这里反正都要初始化，为什么不把这个初始化写在for之前，这样还不用判断了
            - 网友答：没问题啊 这里是for循环里的，每一次循环的是一个epoch而不是一个batch,就算不是random的，一开始的state也要初始化成全0
            - 网友答：这边for循环里的begin_state意思是重置隐状态，而不是仅初始化
            2，老师：如果用的randomiter，前面sequance的信息和当前sequance的信息不是连续的，所以上一个批量的state不该用到这里，
            因为时序上不连续；所以当前批量从头开始是一个新的sequance
            3，老师：如果用的sequanceieter，那么else就不会做重新初始化
            '''
            state = net.begin_state(batch_size=X.shape[0],device=device)
        else:
            # 网友：这里的意思就是说，h0需要初始化，如果当前batch和前一个batch之间不连续的话，h0就不能通用，需要重新初始化
            if isinstance(net, nn.Module) and not isinstance(state, tuple):
                state.detach_()
            else:
                for s in state:
                    s.detach_()
        # y做转置是要把时间序列的维度拉到前面，reshape（-1）是拉成一个向量。网友：Y必须要转置，不然的话预测的概率不是一一对应的
        y = Y.T.reshape(-1)
        # 把数据复制到GPU
        X, y = X.to(device), y.to(device)
        # 做forward函数
        y_hat, state = net(X, state)
        '''
        # 计算loss
        1，老师：这里也可以知道为什么要在RNN里面把所有输出concat起来，这是因为从loss角度讲，这就是一个多分类问题；
        在每一个序列里面任何一个时间步就是一个样本，所以把你拉长拉成一条向量；所以yhat对loss来讲，就是batchsize*timestep个数的样本；
        这里y也用long函数拉成向量了，所以这里就是标准的多分类问题，只是newbatchsize==batchsize*timestep
        
        '''
        l = loss(y_hat, y.long()).mean()
        # 下面这些就是标准的多分类问题，没有别的可说的
        if isinstance(updater, torch.optim.Optimizer):
            # 用了torch的优化器，就要zerograd
            updater.zero_grad()
            l.backward()
            # 在算完backward后，会对梯度裁剪；这里不准梯度超过1；这就是在更新之前做一次剪裁
            grad_clipping(net,1)
            updater.step()
        else:
            l.backward()
            grad_clipping(net,1)
            updater(batch_size=1)
        metric.add(l * y.numel(), y.numel())
    '''
    1，metric[0]/metric[1]:loss除以批量大小，其实就是平均交叉熵
    2，交叉熵做exp，就得到perplexity了
    3，metric[1]/timer.stop()：就是speed
    '''
    return math.exp(metric[0]/metric[1]), metric[1]/timer.stop()
```

```python
# 训练函数支持从零开始或使用高级API实现的循环神经网络模型；之后lstm和GRU可以重用这个函数了
def train_ch8(net, train_iter, vocab, lr, num_epochs, device, use_random_iter=False):
    """训练模型"""
    # 语言模型本质是一个多分类，所以用的是交叉熵
    loss = nn.CrossEntropyLoss()
    # 做动画的东西
    animator = d2l.Animator(xlabel='epoch', ylabel='perplexity', legend=['train'],xlim=[10,num_epochs])    
    # 如果用nnModule就使用torch的优化器；否则使用之前实现过的d2l.sgd优化器
    if isinstance(net, nn.Module):
        updater = torch.optim.SGD(net.parameters(), lr)
    else:
        updater = lambda batch_size: d2l.sgd(net.params, lr, batch_size)
    # predict就是predict_ch8包装一下，就给定prefix就往后预测五十个char
    predict = lambda prefix: predict_ch8(prefix, 50, net, vocab, device)
    # 对每个epoch进行训练
    for epoch in range(num_epochs):
        # 训练一个epoch；返回perplexity和sppeed
        ppl, speed = train_epoch_ch8(net, train_iter, loss, updater, device, use_random_iter)   
        if (epoch + 1) % 10 == 0:
            # 针对每个epoch都会让模型预测一下'time traveller’后面跟的句子，实时看模型的效果
            print(predict('time traveller'))
            animator.add(epoch + 1, [ppl])
    # 训练完成后，打印：困惑度，速度，预测结果
    print(f'困惑度 {ppl:.1f}, {speed:.1f} 标记/秒 {str(device)}')
    print(predict('time traveller'))
    print(predict('traveller'))
```

```python
# 现在我们可以训练循环神经网络模型
num_epochs, lr = 500, 1
train_ch8(net, train_iter, vocab, lr, num_epochs, d2l.try_gpu())
```

- 老师：每次采样都是随机取一个sequence，和上一个批量没什么关系，有更多的随机性；可以看到困惑度会更高并且下降更慢，这是因为每一次state都是打乱的，每一个小批量都是被初始化为0，每次只看一个短的sequence随机性更高
- 老师：看预测的结果，因为训练的样本太小，基本就是记住了很多东西

# 2. RNN的实现(用框架)

```python
# 先把数据加载进来
import torch
from torch import nn
from torch.nn import functional as F
from d2l import torch as d2l

batch_size, num_steps = 32, 35
train_iter, vocab = d2l.load_data_time_machine(batch_size, num_steps)
```

```python
# 定义模型
num_hiddens = 256
# torch的nnmodule中有个RNN的类。传入：输入输出是多少，隐藏层节点数是多少。返回RNN层。
rnn_layer = nn.RNN(len(vocab), num_hiddens)
```

```python
# 使用张量来初始化隐藏状态
state = torch.zeros((1, batch_size, num_hiddens))
state.shape
```

```python
# 通过一个隐藏状态和一个输入，我们可以用更新后的隐藏状态计算输出
X = torch.rand(size=(num_steps, batch_size, len(vocab)))# 初始化输入，方式和从零实现相同；维度为：时间步，批量大小，词典大小
Y, state_new = rnn_layer(X, state)# y维度：时间步，批量数，隐藏层节点数
Y.shape, state_new.shape
```

- 我和网友问：y最后一个维度为啥不是vocab size 而是num_hiddens??
    - 网友答：rnn_layer只包含隐藏的循环层，还需要创建一个单独的输出层
    - 网友答：因为你的rnn的输出num_hiddens
    - 网友答：pytorch的RNN没有实现输出层，所以是256
    - 老师：我们从0实现的RNN包含了输出层，但是torch的RNN是没有输出层的
- 网友问：Y和state的关系是啥
    - 高赞网友答：Y是所有时间步的隐藏层状态，state是最后一个时间步的；state == Y[-1]
    - 注意我：看完(http://t.csdn.cn/crSKs)，结合从0实现RNN，有点理解了。因为RNN及其变形总是依据时间步从T0开始求，然后求到Tt，这样其实隐藏状态只需要保存上一个时间步的就可以了，不需要保存所有时间步的隐藏状态；所以训练完后模型有价值的也是最后一步的隐藏层，故最后一步的隐藏层会和output一起被输出。

```python
# 我们为一个完整的循环神经网络模型定义一个RNNModel类
class RNNModel(nn.Module):
    """循环神经网络模型"""
    def __init__(self, rnn_layer, vocab_size, **kwargs):
        super(RNNModel, self).__init__(**kwargs)
        self.rnn = rnn_layer
        self.vocab_size = vocab_size
        self.num_hiddens = self.rnn.hidden_size
        # 框架rnn后需要输出层，所以可以看到线性层
        if not self.rnn.bidirectional:
            self.num_directions = 1
            self.linear = nn.Linear(self.num_hiddens, self.vocab_size)
        else:
            self.num_directions = 2
            self.linear = nn.Linear(self.num_hiddens * 2, self.vocab_size)  
            
    def forward(self, inputs, state):
        X = F.one_hot(inputs.T.long(), self.vocab_size)
        X = X.to(torch.float32)
        Y, state = self.rnn(X, state)
        # 把y的时间步数和batchsize放在一起作为第一维，后面一个维度即隐藏层大小放在第二维
        output = self.linear(Y.reshape((-1, Y.shape[-1])))
        return output, state

    def begin_state(self, device, batch_size=1):
        # 根据当前rnn是否为LSTM的实例，做出不同返回；这先跳过，说到双向LSTM就懂了
        if not isinstance(self.rnn, nn.LSTM):
            return torch.zeros((self.num_directions * self.rnn.num_layers,
                                batch_size, self.num_hiddens),
                               device=device)
        else:
            return (torch.zeros((self.num_directions * self.rnn.num_layers,
                                 batch_size, self.num_hiddens),
                                device=device), 
                    torch.zeros((self.num_directions * self.rnn.num_layers,
                                 batch_size, self.num_hiddens), 
                                device=device))
```

- 老师：可以看到用了RNNmodule的好处就是不用管forward了，但是训练模型整体上和rnnscratch没太多区别

```python
# 用一个具有随即权重的模型进行预测：看没训练的时候predict是怎样
device = d2l.try_gpu()
net = RNNModel(rnn_layer, vocab_size=len(vocab))
net = net.to(device)
d2l.predict_ch8('time traveller', 10, net, vocab, device)
```

```python
# 使用高级API训练模型
num_epochs, lr = 500, 1
d2l.train_ch8(net, train_iter, vocab, lr, num_epochs, device)
```
